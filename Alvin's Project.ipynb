# Cleaning the data

We have some rows with no District... we can't do much with them. So we're just going to drop them. There aren't that many, though (only 47 in our table with about 6 million rows), so it shouldn't have an impact. Also, remove all rows which are in district 31... this district isn't part of the regular list on the website and i can't find any info on them

```python
original_num_rows = len(crime)
crime = crime[ crime['District'].isnull() == False ] # remove the rows with null district
crime
```

```python
assert len(crime) == original_num_rows - num_row_null_district
```

Now, we want to see which primary types of crimes have happened the most and grab the top two

```python
crime = crime[ crime['District'] <= 25 ]
crime['Primary Type'].value_counts()
```

We can see that the top two types are Theft (1436417
) and Battery (1245721
).
Therefore, we know want to see how many of each type actually concluded with an arrest.

```python
print(crime[(crime['Primary Type']=='THEFT') & (crime['Arrest']==True)].shape)
print(crime[(crime['Primary Type']=='BATTERY') & (crime['Arrest']==True)].shape)
#print(crime[(crime['Primary Type']=='CRIMINAL DAMAGE') & (crime['Arrest']==True)].shape)
```

Now, we want to see the data in regards to our other possible variable (District). So we get the number of crimes for district first.

```python
crime['District'].value_counts()
```

We can see that the top 2 districts with crimes are Districts 8.0 and 11.0. Therefore, we want to see out of all these reported crimes in these 2 districts, how many actually resulted in an arrest? 

```python
print(crime[(crime['District']==8.0) & (crime['Arrest']==True)].shape)
print(crime[(crime['District']==11.0) & (crime['Arrest']==True)].shape)
```
